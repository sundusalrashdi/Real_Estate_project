```python
import pandas as pd
from io import StringIO

# Read the first file (synthetic data)
file_path1 = r"germany_property_data.csv"
df1 = pd.read_csv(file_path1)

# Read the second file (cleaned real data)
file_path2 = r"egypt_property_data.csv"
df2 = pd.read_csv(file_path2)


# Concatenate both DataFrames on the common columns
merged_df = pd.concat([df1, df2], ignore_index=True)

# Save the merged DataFrame
merged_df.to_csv("merged_real_estate.csv", index=False)
print(" Merged file saved as merged_real_estate.csv")

# Display the first few rows
merged_df 
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
columns_to_encode = [
    'address', 'city', 'state', 'country', 'propertyType' 
]
for column in columns_to_encode:
    if column in merged_df.columns:
        merged_df[column] = label_encoder.fit_transform(merged_df[column])
```

```python
merged_df
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt


X  = merged_df.drop(columns = 'mostRecentPriceAmount')
y = merged_df['mostRecentPriceAmount']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 1. Linear Regression
linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)
linear_preds = linear_model.predict(X_test_scaled)

linear_mse = mean_squared_error(y_test, linear_preds)
linear_r2 = r2_score(y_test, linear_preds)
print("Linear Regression MSE:", linear_mse)
print("Linear Regression R² Score:", linear_r2)


```

```python
# 2. Decision Tree Regressor
# ---------------------------
tree_model = DecisionTreeRegressor()
tree_model.fit(X_train, y_train)
tree_preds = tree_model.predict(X_test)

tree_mse = mean_squared_error(y_test, tree_preds)
tree_r2 = r2_score(y_test, tree_preds)
print("Decision Tree MSE:", tree_mse)
print("Decision Tree R² Score:", tree_r2)
```

```python
# 3. Random Forest Regressor
# ---------------------------
forest_model = RandomForestRegressor()
forest_model.fit(X_train, y_train)
forest_preds = forest_model.predict(X_test)

forest_mse = mean_squared_error(y_test, forest_preds)
forest_r2 = r2_score(y_test, forest_preds)
print("Random Forest MSE:", forest_mse)
print("Random Forest R² Score:", forest_r2)
```

```python
plt.figure(figsize=(18, 5))

# 1. Linear Regression
plt.subplot(1, 3, 1)
plt.scatter(y_test, linear_preds, alpha=0.7, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title('Linear Regression\nTrue vs Predicted')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
```

```python
# 2. Decision Tree Regressor
plt.subplot(1, 3, 2)
plt.scatter(y_test, tree_preds, alpha=0.7, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title('Decision Tree\nTrue vs Predicted')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
```

```python
# 3. Random Forest Regressor
plt.subplot(1, 3, 3)
plt.scatter(y_test, forest_preds, alpha=0.7, color='orange')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title('Random Forest\nTrue vs Predicted')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')

plt.tight_layout()
plt.show() 
```
